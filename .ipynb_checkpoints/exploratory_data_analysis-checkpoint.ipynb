```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline

# helper functions
def preprocess_multiple_choice(ser,sep):
    """
    return preprocessed dataframe for multiple choice formats.
    
    a multiple choice format frequently written with the
    choices separated by an arbitrary character like the 
    following format:
    
    "choice1;choice2;choice3;"
    
    This function accepts condensed data and expands it to 
    form a sparse dataframe.
    
    Parameters
    ----------
    ser : array-like
        A Series or List containing multiple choices.
    
    sep : string
        A string that separates the multiple choices inside 
        each ser.
    
    Examples
    --------
    >> s = pd.Series(['A;B', 'B;C', 'A;C'])
    >> preprocess_multiple_choice(s)
    |    | 0    | 1    | 2    |
    |---:|:-----|------|------|
    |  0 | A    | B    | 0    |
    |  1 | 0    | B    | C    |
    |  2 | A    | 0    | C    |
    """
    #get the index of the largest length.
    #then build a reference list to guide where to put subsequent values
    idx = ser.str.len().sort_values(ascending=False).index[0]
    referencelist = ser[idx].split(sep)
    #place for the would-be returned df.
    dict_for_df = {}
    for index, splittedlist in ser.items():
        id = index
        sparselist = pd.Series(np.repeat(np.nan, len(referencelist)))
        if pd.isna(splittedlist):
            dict_for_df[id] = sparselist
        else:
            splittedlist = splittedlist.split(sep)
            for word in splittedlist:
                wordindex = referencelist.index(word)
                sparselist[wordindex] = word
            dict_for_df[id] = sparselist
    return pd.DataFrame.from_dict(dict_for_df, orient='index')

def quickbar(s):
    """
    Just a shorthand to build a bar plot using the value counted
    result of a series.
    
    because this is a horizontal bar, the index is provided as 
    the argument for y, while the values are provided as
    the argument for x.
    """
    plt.barh(s.value_counts().index, s.value_counts().values)
    plt.title(df_questions[(df_questions['qname'] == s.name)]['question'].values[0])

# def quickbar_mc(data, title=None):
#     """
#     Just a shorthand to build a bar plot using the value counted
#     result of a series, similar to quickbar. This method is for
#     unprocessed multiple choice data.
    
#     because this is a horizontal bar, the index is provided as 
#     the argument for y, while the values are provided as
#     the argument for x.
#     """
#     sparse = preprocess_multiple_choice(data,';')
#     sparse_vc = count_val(sparse)
#     plt.barh(sparse_vc.index, sparse_vc.values)
#     if title:
#         plt.title(title)
#     else:
#         plt.title(df_questions[(df_questions['qname'] == data.name)]['question'].values[0])
        
def quickbar_mc(ser, title=None):
    """
    New version, faster and less buggy.
    """
    _ = ser.str.split(';', expand=True).stack().value_counts()
    plt.barh(_.index, _.values)
    if title:
        plt.title(title)
    else:
        plt.title(df_questions[(df_questions['qname'] == ser.name)]['question'].values[0])
    
def count_ma(df, normalize=False):
    if normalize == False:
        count_result = df.count()
    else:
        count_result = df.count() / df.count().sum()
    return pd.Series(count_result.values, index=df.mode().iloc[0].values)

def count_val(data, normalize=False):
    if isinstance(data, pd.Series):
        return data.value_counts(normalize=normalize).sort_values()
    elif isinstance(data, pd.DataFrame):
        return count_ma(data, normalize=normalize).sort_values()
```

```python
#We concatenate so the position could be aligned.
language = pd.concat([df['LanguageWantToWorkWith'].str.split(';', expand=True).stack().value_counts().rename('Want'),
                      df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().value_counts().rename('Have')], axis=1)

language['Difference'] = language['Want'] - language['Have']
language.sort_values('Difference', inplace=True)

pos = np.arange(len(language))
height = 0.3
f, ax = plt.subplot_mosaic([[0,0,1]], figsize=(12,9))
ax[0].barh(pos-height/2, language['Want'], height=height, label='I Want to Work With This Tech')
ax[0].barh(pos+height/2, language["Have"], height=height, label='I Have Worked With This Tech')
ax[0].legend()
ax[0].set_yticks(pos, language.index)
ax[1].stem(language.index, language['Difference'], orientation='horizontal')
ax[1].tick_params(left=False,labelleft=False,right=True,labelright=True)
plt.show()
```
